```python
sns.barplot(data=client_data, x="churn", y="net_margin")
plt.show()
```


#### Net Margin Comparison:
Churn = 0 (Blue Bar): The average net margin for customers who did not churn is lower.
Churn = 1 (Orange Bar): The average net margin for customers who did churn is higher.

#### Implications:
Higher Margin, Higher Churn: The chart indicates that customers with a higher net margin are more likely to churn. This could be a concerning sign because losing high-margin customers can significantly impact the company's profitability.
Lower Margin, Lower Churn: Conversely, customers with a lower net margin are less likely to churn, which might mean that the company is retaining less profitable customers while losing more profitable ones.

#### Error Bars:
The black lines on top of the bars represent the error bars, which show the variability of the data. The larger the error bar, the more variability there is in the net margin among those customers.
Interpretation:

#### Customer Profitability and Churn: 
The chart suggests a counterintuitive trend where the company is losing more profitable customers (those with higher net margins). This might indicate underlying issues such as dissatisfaction among high-value customers, competitive pressures, or inadequate retention strategies for this segment.

#### Strategic Focus: 
The company may need to investigate why high-margin customers are churning and take steps to address the causes, such as improving customer service, enhancing product offerings, or offering incentives to retain these valuable customers.

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
churn = client_data[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert price_date to datetime format
price_data['price_date'] = pd.to_datetime(price_data['price_date'])

# Aggregate data by date (mean for each date)
aggregated_data = price_data.groupby('price_date').mean()

plt.figure(figsize=(10, 6))
plt.plot(aggregated_data.index, aggregated_data['price_off_peak_var'], label='Off-Peak Variable')
plt.plot(aggregated_data.index, aggregated_data['price_peak_var'], label='Peak Variable')
plt.plot(aggregated_data.index, aggregated_data['price_mid_peak_var'], label='Mid-Peak Variable')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Average Variable Price Variations Over Time')
plt.legend()
plt.show()
```

#### Pricing Stability: 
The stability in prices suggests that PowerCoâ€™s variable pricing structure does not fluctuate significantly over short periods (e.g., within a year). This might indicate a stable market or a deliberate pricing strategy to maintain customer predictability.

#### Unexpected Price Hierarchy: 
Typically, one might expect peak times to be more expensive, but in this case, Off-Peak Variable prices are higher than Peak Variable prices. This could be due to specific contract structures, market demand, or other factors influencing pricing.

#### Strategic Considerations: 
If PowerCo is looking to adjust pricing to influence customer behavior (e.g., shifting usage to off-peak times), they might need to re-evaluate these price levels, as the current structure might not be providing strong incentives for customers to change their usage patterns.
